- ### Subtask 3.5: Find the best directors

    1. Group the dataframe using the `director_name` column.
    2. Find out the top 10 directors for whom the mean of `imdb_score` is the highest and store them in a new dataframe `top10director`. 

```python
temp = movies.groupby( 'director_name', axis = 0 )
temp = temp.first()


temp = temp.reset_index()
imdb_score_mean_list = []

for i in temp['director_name']:
    mean = 0
    
    l = list( movies[ movies['director_name'] == i ]['imdb_score'] )
    
    if( len(l) == 1 ):
        mean = l[0]
    else:
        for j in l:
            mean = mean + j
        
        mean = mean/len(l)
        
    imdb_score_mean_list.append( mean )
    
temp['mean']= imdb_score_mean_list

temp.sort_values( by = ['mean'], ascending = False, inplace= True )

top10director = temp.head( 10 )
top10director
```

-  ### Subtask 3.6: Find popular genres

You might have noticed the `genres` column in the dataframe with all the genres of the movies seperated by a pipe (`|`). Out of all the movie genres, the first two are most significant for any film.

1. Extract the first two genres from the `genres` column and store them in two new columns: `genre_1` and `genre_2`. Some of the movies might have only one genre. In such cases, extract the single genre into both the columns, i.e. for such movies the `genre_2` will be the same as `genre_1`.
2. Group the dataframe using `genre_1` as the primary column and `genre_2` as the secondary column.
3. Find out the 5 most popular combo of genres by finding the mean of the gross values using the `gross` column and store them in a new dataframe named `PopGenre`.

```python
# Write your code for extracting the first two genres of each movie here

genre_1_list = []
genre_2_list = []

for i in movies['genres']:
    temp = i.split('|')
    
    if( len( temp ) == 1 ):
        genre_1_list.append( temp[0] )
        genre_2_list.append( temp[0] )
    else:
        genre_1_list.append( temp[0] )
        genre_2_list.append( temp[1] )
        
temp = movies.copy()

temp['genre_1'] = genre_1_list
temp['genre_2'] = genre_2_list

temp
```

```python
# Write your code for grouping the dataframe here

movies_by_segment = temp.groupby( [ 'genre_1', 'genre_2' ], axis = 0 )
movies_by_segment = movies_by_segment.first()

movies_by_segment = movies_by_segment.reset_index()
movies_by_segment
```

```python
# Write your code for getting the 5 most popular combo of genres here
gross_mean_dict = { 'genre_1' : movies_by_segment['genre_1']
                  ,'genre_2' : movies_by_segment['genre_2'] }

gross_mean_dict['gross_sum'] = [0] * len( gross_mean_dict['genre_1'] )
gross_mean_dict['gross_count'] = [0] * len( gross_mean_dict['genre_1'] )

gross_mean_dict = pd.DataFrame( gross_mean_dict )

# print( f"temp.shape = {  temp.shape}" )
# temp.index = list( range( 0, 3784 ) )

# print( temp['genre_1'] )
# print()
# print( temp['genre_2'] )
# print()
# print( temp )

count = 0
for i in range( len( temp ) ):
    #print(i,  ( temp['genre_1'][i] in list( gross_mean_dict['genre_1'] ) ) and ( temp['genre_2'][i] in list( gross_mean_dict['genre_2'] ) ) )
    
    if( ( temp['genre_1'][i] in list( gross_mean_dict['genre_1'] ) ) and ( temp['genre_2'][i] in list( gross_mean_dict['genre_2'] ) ) ):
        gross_index = gross_mean_dict[ ( ( gross_mean_dict['genre_1'] == temp['genre_1'][i] ) & ( gross_mean_dict['genre_2'] == temp['genre_2'][i] ) ) ].index
        
        #print( f"gross_index = { gross_index }" )
        
        gross_mean_dict['gross_sum'][gross_index] = gross_mean_dict['gross_sum'][gross_index] + temp['gross'][i]
        gross_mean_dict['gross_count'][gross_index] = gross_mean_dict['gross_count'][gross_index] + 1
        
        count = count + 1
```

```python
gross_mean_dict['gross_mean'] = gross_mean_dict['gross_sum'] / gross_mean_dict['gross_count']

gross_mean_dict.sort_values( by = ['gross_mean'], ascending = False, inplace= True )

PopGenre = gross_mean_dict.head( 5 )
PopGenre
```

```python
# Write your code for grouping the combined dataframe here

#3. Group the combined dataframe using the `actor_1_name` column.

temp = Combined.groupby( 'actor_1_name', axis = 0 )
temp = temp.first()


temp = temp.reset_index()
temp
```

```python
# Write the code for finding the mean of critic reviews and audience reviews here
#4. Find the mean of the `num_critic_for_reviews` and `num_user_for_review` and identify the actors which have the highest mean.

critic_dataframe = { 'actor_name' : temp['actor_1_name'] }

critic_dataframe['critic_sum'] = [0] * 3
critic_dataframe['critic_count'] = [0] * 3


critic_dataframe

for i in range( len(Combined) ):
    actor_name = Combined['actor_1_name'][i]
    
    if( actor_name == 'Brad Pitt' ):
        critic_dataframe['critic_sum'][0] =  critic_dataframe['critic_sum'][0] + Combined['num_critic_for_reviews'][i]
        
        critic_dataframe['critic_count'][0] =  critic_dataframe['critic_count'][0] + 1
        
    elif( actor_name == 'Leonardo DiCaprio' ):
        critic_dataframe['critic_sum'][1] =  critic_dataframe['critic_sum'][1] + Combined['num_critic_for_reviews'][i]
        
        critic_dataframe['critic_count'][1] =  critic_dataframe['critic_count'][1] + 1
        
    elif( actor_name == 'Meryl Streep' ):
        critic_dataframe['critic_sum'][2] =  critic_dataframe['critic_sum'][2] + Combined['num_critic_for_reviews'][i]
        
        critic_dataframe['critic_count'][2] =  critic_dataframe['critic_count'][2] + 1
        
critic_dataframe = pd.DataFrame( critic_dataframe )
critic_dataframe['mean'] = critic_dataframe['critic_sum'] / critic_dataframe['critic_count']



critic_dataframe.sort_values( by = ['mean'], ascending = False, inplace= True )

critic_dataframe.index = list( range( len( critic_dataframe ) ) )

print( critic_dataframe['actor_name'][0] )
```
